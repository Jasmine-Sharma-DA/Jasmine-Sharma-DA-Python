**Remove columns that are 90% empty**

```python
#taking only 10 % of the data 
threshold=len(df)*0.1
threshold
```

```python
#to remove the rows and columns having null values
#axis=1 means remove row, and axis=0 means remove column
df.dropna(thresh=threshold,axis=1,inplace=True)
```

```python
print(df.isnull().sum())
```

**Data Manipulation**

```python
#how to average out and fill in the null values
def impute_median(series):
    return series.fillna(series.median)
```

```python
#filling the data with the impute_median 
df.Rating=df['Rating'].transform(impute_median)
```

```python
df.isnull().sum()
```

```python
#filled the null value with mode
print(df['Type'].mode())
print(df['Current Ver'].mode())
print(df['Android Ver'].mode())
```

```python
df['Type'].fillna(str(df['Type'].mode().values[0]),inplace=True)
df['Current Ver'].fillna(str(df['Current Ver'].mode().values[0]),inplace=True)
df['Android Ver'].fillna(str(df['Android Ver'].mode().values[0]),inplace=True)
```

```python
df.isnull().sum()
```

**Data Visualizaion**

```python
# Step 1: Remove non-numeric characters from the 'Installs' column
df['Installs'] = df['Installs'].str.replace('[^\d]', '', regex=True)
df['Price'] = df['Price'].str.replace('[^\d]', '', regex=True)
df['Reviews'] = df['Reviews'].str.replace('[^\d]', '', regex=True)

# Step 2: Convert the cleaned 'Installs' column to a numeric data type
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

# Step 3: Calculate the mean
grp = df.groupby('Category')
x = grp['Installs'].agg(np.mean)
y = grp['Price'].agg(np.sum)
z = grp['Reviews'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
plt.figure(figsize=(16,5))
plt.plot(x, 'ro', color='b') #ro =rount circles and b= blue
plt.xticks(rotation=90)
plt.title('Category vs Installs')
plt.xlabel('Categories  ----->')
plt.ylabel('Installs ----->')
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot( y, 'r--', color='r') #r-- = dotted line and r= red
plt.xticks(rotation=90)
plt.title('Category vs Pricing')
plt.xlabel('Categories  ----->')
plt.ylabel('Prices ----->')
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot( z, 'g^', color='g') #g^ = ^ and g= green
plt.xticks(rotation=90)
plt.title('Category vs Reviews')
plt.xlabel('Categories  ----->')
plt.ylabel('Reviews ----->')
plt.show()
```
