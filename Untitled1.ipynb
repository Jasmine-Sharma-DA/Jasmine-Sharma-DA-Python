```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("data.xlsx")
```

```python
data.info()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
#filled the null value with mode
print(data['Restaurant Name'].mode())
print(data['Cuisines'].mode())
```

```python
data['Restaurant Name'].fillna(str(data['Restaurant Name'].mode().values[0]),inplace=True)
data['Cuisines'].fillna(str(data['Cuisines'].mode().values[0]),inplace=True)
```

```python
data.isnull().sum()
```

```python
unnecessary_columns = ['Locality Verbose','Longitude','Latitude']
data.drop(unnecessary_columns,axis=1,inplace=True)
```

```python
data.isnull().sum()
```

**Preliminary Report and Exploratory Data Analysis (EDA)**

```python
# Count restaurants by city
city_counts = data['City'].value_counts()

# Plot geographical distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=city_counts.index, y=city_counts.values, palette='viridis')
plt.xticks(rotation=90)
plt.title('Number of Restaurants by City')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Plot rating distribution
plt.figure(figsize=(8, 6))
sns.histplot(data['Aggregate rating'], bins=20, kde=True)
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count restaurants by franchise
franchise_counts = data['Restaurant Name'].value_counts()

# Plot franchise presence
plt.figure(figsize=(10, 6))
sns.barplot(x=franchise_counts.index, y=franchise_counts.values, palette='coolwarm')
plt.xticks(rotation=90)
plt.title('National Presence of Franchises')
plt.xlabel('Franchise')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Calculate ratio
table_booking_ratio = data['Has Table booking'].value_counts(normalize=True) * 100

# Plot table booking ratio
plt.figure(figsize=(6, 6))
table_booking_ratio.plot.pie(autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
plt.title('Table Booking Availability')
plt.ylabel('')
plt.show()
```

```python
# Calculate percentage
online_delivery_percentage = data['Has Online delivery'].value_counts(normalize=True) * 100

# Plot online delivery percentage
plt.figure(figsize=(6, 6))
online_delivery_percentage.plot.pie(autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Online Delivery Availability')
plt.ylabel('')
plt.show()
```

```python
# Calculate mean votes for delivery and non-delivery
votes_delivery = data[data['Has Online delivery'] == 'Yes']['Votes'].mean()
votes_no_delivery = data[data['Has Online delivery'] == 'No']['Votes'].mean()

print(f'Average votes for delivery: {votes_delivery}')
print(f'Average votes for non-delivery: {votes_no_delivery}')
```
